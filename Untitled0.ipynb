```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from datetime import datetime, timedelta
import warnings
```

```python
# For better visualization
plt.style.use('seaborn-v0_8-whitegrid')
sns.set(style="whitegrid")
warnings.filterwarnings('ignore')
```

```python
plt.rcParams["figure.figsize"] = (12, 8)
```

```python
url = "https://covid.ourworldindata.org/data/owid-covid-data.csv"
df = pd.read_csv(url)
```

```python
print(f"Dataset shape: {df.shape}")
print(f"\nTime range: {df['date'].min()} to {df['date'].max()}")
print(f"\nNumber of locations: {df['location'].nunique()}")
```

```python
print("First 5 rows of the dataset:")
df.head()
```

```python
print("Column names in the dataset:")
print(df.columns.tolist())
```

```python
print("\nData types and missing values:")
df.info()
```

```python
missing_values = df[['iso_code', 'continent', 'location', 'date', 'total_cases',
                     'new_cases', 'total_deaths', 'new_deaths',
                     'total_vaccinations', 'people_vaccinated']].isnull().sum()

print("\nMissing values in key columns:")
print(missing_values)
```

```python
# Convert date to datetime format
df['date'] = pd.to_datetime(df['date'])

# Create a new column for month-year
df['month_year'] = df['date'].dt.strftime('%b %Y')

# Select countries of interest for detailed analysis
countries_of_interest = ['United States', 'India', 'Brazil', 'United Kingdom', 'Russia', 'South Africa', 'Kenya']

# Filter data for these countries
countries_df = df[df['location'].isin(countries_of_interest)]

# Get data for continents for broader analysis
continents_df = df[df['continent'].notna()]

# Get latest data for each country (for maps and current status)
latest_date = df['date'].max()
latest_data = df[df['date'] == latest_date]

# Sort by total cases descending
latest_data_sorted = latest_data.sort_values('total_cases', ascending=False)

print(f"Latest data date: {latest_date}")
print(f"Number of countries in latest data: {latest_data['location'].nunique()}")
```

```python
# Calculate global totals by date
global_data = df.groupby('date')[['new_cases', 'new_deaths', 'total_cases', 'total_deaths']].sum().reset_index()

# Plot global cases and deaths over time
fig, ax = plt.subplots(2, 1, figsize=(14, 12))

# Total cases
ax[0].plot(global_data['date'], global_data['total_cases']/1000000, color='steelblue', linewidth=2)
ax[0].set_title('Global COVID-19 Total Cases Over Time', fontsize=16)
ax[0].set_ylabel('Total Cases (millions)', fontsize=12)
ax[0].grid(True, alpha=0.3)

# Total deaths
ax[1].plot(global_data['date'], global_data['total_deaths']/1000000, color='darkred', linewidth=2)
ax[1].set_title('Global COVID-19 Total Deaths Over Time', fontsize=16)
ax[1].set_xlabel('Date', fontsize=12)
ax[1].set_ylabel('Total Deaths (millions)', fontsize=12)
ax[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
 #Daily new cases and deaths
fig, ax = plt.subplots(2, 1, figsize=(14, 12))

# 7-day moving average for smoother visualization
global_data['new_cases_7day_avg'] = global_data['new_cases'].rolling(window=7).mean()
global_data['new_deaths_7day_avg'] = global_data['new_deaths'].rolling(window=7).mean()

# New cases
ax[0].plot(global_data['date'], global_data['new_cases_7day_avg']/1000, color='steelblue', linewidth=2)
ax[0].set_title('Global COVID-19 Daily New Cases (7-day moving average)', fontsize=16)
ax[0].set_ylabel('New Cases (thousands)', fontsize=12)
ax[0].grid(True, alpha=0.3)

# New deaths
ax[1].plot(global_data['date'], global_data['new_deaths_7day_avg']/1000, color='darkred', linewidth=2)
ax[1].set_title('Global COVID-19 Daily New Deaths (7-day moving average)', fontsize=16)
ax[1].set_xlabel('Date', fontsize=12)
ax[1].set_ylabel('New Deaths (thousands)', fontsize=12)
ax[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Get top 10 countries by total cases
top_cases = latest_data_sorted[latest_data_sorted['location'].isin(df['location'].unique()) &
                               ~latest_data_sorted['location'].isin(['World', 'Europe', 'Asia', 'North America',
                                                                   'South America', 'Africa', 'Oceania'])].head(10)

# Get top 10 countries by total deaths
top_deaths = latest_data_sorted.sort_values('total_deaths', ascending=False)
top_deaths = top_deaths[top_deaths['location'].isin(df['location'].unique()) &
                         ~top_deaths['location'].isin(['World', 'Europe', 'Asia', 'North America',
                                                     'South America', 'Africa', 'Oceania'])].head(10)

# Create subplots
fig, ax = plt.subplots(2, 1, figsize=(12, 14))

# Plot top countries by cases
ax[0].barh(top_cases['location'][::-1], top_cases['total_cases'][::-1]/1000000, color='steelblue')
ax[0].set_title('Top 10 Countries by Total COVID-19 Cases', fontsize=16)
ax[0].set_xlabel('Total Cases (millions)', fontsize=12)
for i, v in enumerate(top_cases['total_cases'][::-1]/1000000):
    ax[0].text(v + 0.1, i, f"{v:.2f}M", va='center')

# Plot top countries by deaths
ax[1].barh(top_deaths['location'][::-1], top_deaths['total_deaths'][::-1]/1000, color='darkred')
ax[1].set_title('Top 10 Countries by Total COVID-19 Deaths', fontsize=16)
ax[1].set_xlabel('Total Deaths (thousands)', fontsize=12)
for i, v in enumerate(top_deaths['total_deaths'][::-1]/1000):
    ax[1].text(v + 10, i, f"{v:.1f}K", va='center')

plt.tight_layout()
plt.show()
```

```python
countries_data = df[df['location'].isin(countries_of_interest)]

# Pivot the data for easier plotting
cases_pivot = countries_data.pivot(index='date', columns='location', values='total_cases')
deaths_pivot = countries_data.pivot(index='date', columns='location', values='total_deaths')

# Plot total cases
plt.figure(figsize=(14, 8))
for country in countries_of_interest:
    if country in cases_pivot.columns:
        plt.plot(cases_pivot.index, cases_pivot[country]/1000000, linewidth=2, label=country)

plt.title('Total COVID-19 Cases by Country Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Cases (millions)', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Plot total deaths
plt.figure(figsize=(14, 8))
for country in countries_of_interest:
    if country in deaths_pivot.columns:
        plt.plot(deaths_pivot.index, deaths_pivot[country]/1000, linewidth=2, label=country)

plt.title('Total COVID-19 Deaths by Country Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Deaths (thousands)', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Calculate new cases and deaths per million for normalized comparison
countries_data_norm = countries_data.copy()

# 7-day moving average
for country in countries_of_interest:
    country_data = countries_data_norm[countries_data_norm['location'] == country]
    countries_data_norm.loc[countries_data_norm['location'] == country, 'new_cases_smoothed_per_million'] = \
        country_data['new_cases_per_million'].rolling(window=7).mean()
    countries_data_norm.loc[countries_data_norm['location'] == country, 'new_deaths_smoothed_per_million'] = \
        country_data['new_deaths_per_million'].rolling(window=7).mean()

# Pivot for easier plotting
cases_per_million_pivot = countries_data_norm.pivot(index='date', columns='location', values='new_cases_smoothed_per_million')
deaths_per_million_pivot = countries_data_norm.pivot(index='date', columns='location', values='new_deaths_smoothed_per_million')

# Plot new cases per million
plt.figure(figsize=(14, 8))
for country in countries_of_interest:
    if country in cases_per_million_pivot.columns:
        plt.plot(cases_per_million_pivot.index, cases_per_million_pivot[country], linewidth=2, label=country)

plt.title('Daily New COVID-19 Cases per Million by Country (7-day moving avg)', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('New Cases per Million', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Calculate death rate (total_deaths / total_cases) for each country
countries_data['death_rate'] = countries_data['total_deaths'] / countries_data['total_cases'] * 100

# Get the latest death rate for each country
latest_death_rate = countries_data[countries_data['date'] == countries_data['date'].max()]
latest_death_rate = latest_death_rate[['location', 'death_rate']].sort_values('death_rate', ascending=False)

# Plot death rate for our countries of interest
plt.figure(figsize=(12, 8))
latest_death_rate_filtered = latest_death_rate[latest_death_rate['location'].isin(countries_of_interest)]
sns.barplot(x='location', y='death_rate', data=latest_death_rate_filtered, palette='coolwarm')
plt.title('COVID-19 Death Rate by Country (Deaths/Cases %)', fontsize=16)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Death Rate (%)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)

# Add text labels above bars
for i, v in enumerate(latest_death_rate_filtered['death_rate']):
    plt.text(i, v + 0.1, f"{v:.2f}%", ha='center')

plt.tight_layout()
plt.show()
```

```python
# Filter data for vaccination analysis
vac_data = df[df['location'].isin(countries_of_interest) & df['total_vaccinations'].notna()]

# Plot total vaccinations over time
vac_pivot = vac_data.pivot(index='date', columns='location', values='total_vaccinations')

plt.figure(figsize=(14, 8))
for country in countries_of_interest:
    if country in vac_pivot.columns:
        plt.plot(vac_pivot.index, vac_pivot[country]/1000000, linewidth=2, label=country)

plt.title('Total COVID-19 Vaccine Doses Administered by Country', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Vaccine Doses (millions)', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Plot vaccination percentage for countries of interest
vac_percent_pivot = vac_data.pivot(index='date', columns='location', values='people_vaccinated_per_hundred')

plt.figure(figsize=(14, 8))
for country in countries_of_interest:
    if country in vac_percent_pivot.columns:
        plt.plot(vac_percent_pivot.index, vac_percent_pivot[country], linewidth=2, label=country)

plt.title('Percentage of Population Vaccinated by Country', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Population Vaccinated (%)', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.ylim(0, 100)  # Set y-axis from 0 to 100%
plt.tight_layout()
plt.show()
```

```python
# Latest vaccination percentages - comparing partial vs. full vaccination
latest_vac = df[df['date'] == df['date'].max()]
latest_vac = latest_vac[latest_vac['location'].isin(countries_of_interest)]
latest_vac = latest_vac[['location', 'people_vaccinated_per_hundred', 'people_fully_vaccinated_per_hundred']].sort_values('people_fully_vaccinated_per_hundred', ascending=False)

# Convert to long format for grouped bar chart
latest_vac_long = pd.melt(latest_vac, id_vars=['location'],
                          value_vars=['people_vaccinated_per_hundred', 'people_fully_vaccinated_per_hundred'],
                          var_name='vaccination_status', value_name='percentage')

# Replace variable names for better labels
latest_vac_long['vaccination_status'] = latest_vac_long['vaccination_status'].replace({
    'people_vaccinated_per_hundred': 'At Least One Dose',
    'people_fully_vaccinated_per_hundred': 'Fully Vaccinated'
})

# Plot grouped bar chart
plt.figure(figsize=(14, 8))
sns.barplot(x='location', y='percentage', hue='vaccination_status', data=latest_vac_long, palette='viridis')
plt.title('COVID-19 Vaccination Status by Country (Latest Data)', fontsize=16)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Population Percentage (%)', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Vaccination Status')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 7. Key Insights and Findings

Based on our analysis of the COVID-19 global data, here are the key insights:

1. **Global Trends:** The COVID-19 pandemic has seen multiple waves of infections globally, with the steepest increases occurring during major variant outbreaks. Total cases reached into the hundreds of millions, with deaths in the millions worldwide.

2. **Country Comparisons:**
   - The United States, India, and Brazil consistently showed the highest absolute number of cases and deaths.
   - When normalized by population, smaller countries often showed higher per-capita impacts.
   - Death rates varied significantly between countries, suggesting differences in healthcare capacity, reporting methods, demographics, and intervention strategies.

3. **Vaccination Progress:**
   - Vaccination rollouts began at different times across countries and progressed at varying rates.
   - High-income countries generally achieved higher vaccination rates more quickly.
   - There's a notable gap between countries with high vaccination rates (>70%) and those with lower rates (<30%).

4. **Relationship Between Vaccination and Cases/Deaths:**
   - Countries with higher vaccination rates generally experienced lower death rates in later waves.
   - Vaccination impact is visible in the decoupling of case and death rates in highly vaccinated populations.

5. **Regional Differences:**
   - Different continents experienced peak waves at different times.
   - Regional cooperation and shared borders influenced the spread patterns.

These insights demonstrate the varied impacts of COVID-19 globally and the importance of public health measures, including vaccination campaigns, in mitigating the pandemic's effects.
